```python
import matplotlib.pyplot as plt
import random

# Vertices of an equilateral triangle
vertices = [(0, 0), (1, 0), (0.5, 0.9)]  # (x, y) coordinates of the vertices

# Function to check whether a point is inside the triangle
def isInside(x1, y1, x2, y2, x3, y3, x, y):
    def area(x1, y1, x2, y2, x3, y3):
        return abs((x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2)) / 2.0)

    A = area(x1, y1, x2, y2, x3, y3)
    A1 = area(x, y, x2, y2, x3, y3)
    A2 = area(x1, y1, x, y, x3, y3)
    A3 = area(x1, y1, x2, y2, x, y)

    return A == A1 + A2 + A3

# Get a valid seed point from the user
while True:
    try:
        seed_x = float(input("Enter the x-coordinate of the seed point: "))
        seed_y = float(input("Enter the y-coordinate of the seed point: "))
        if isInside(0, 0, 1, 0, 0.5, 0.9, seed_x, seed_y):
            print("Valid seed point entered.")
            break
        else:
            print("The point is not inside the triangle. Please try again.")
    except ValueError:
        print("Invalid input. Please enter numerical values.")

# Starting point
seed = (seed_x, seed_y)

# Initialize points list with the seed
points = [seed]

# Number of steps
num_steps = int(input("Enter the number of steps: "))

# Chaos game algorithm
for i in range(num_steps):
    # Choose a random vertex
    next_vertex = vertices[random.randint(0, 2)]
    
    # Calculate the midpoint between the current point and the chosen vertex
    current_point = points[-1]
    next_point = ((current_point[0] + next_vertex[0]) / 2, (current_point[1] + next_vertex[1]) / 2)
    
    # Append the new point to the points list
    points.append(next_point)

# Plot the points
def plot_solution(points):
    plt.scatter([p[0] for p in points[12:]], [p[1] for p in points[12:]], s=0.1)
    plt.title("Sierpinski Triangle")
    plt.show()

plot_solution(points)
```

```python
import matplotlib.pyplot as plt
import random
import math

# 1. Define the Vertices of the Triangle:
vertices = [(0, 0), (1, 0), (0.5, math.sqrt(3)/2)]  # Classic coordinates for equilateral triangle

# 2. Get User Input for Number of Steps (Iterations):
num_steps = int(input("Enter the number of steps: "))

# 3. Choose a Random Starting Point (Seed):
#    For simplicity, we'll start at the first vertex.  A random point inside is better.
start_x = 0.0
start_y = 0.0
seed = (start_x, start_y)

# 4. Initialize the List of Points:
points = [seed]

# 5. Chaos Game Algorithm:
for i in range(num_steps):
    # a. Choose a random vertex:
    random_vertex = vertices[random.randint(0, 2)]  # 0, 1, or 2

    # b. Calculate the midpoint between the current point and the random vertex:
    current_x, current_y = points[-1]  # Get the last point added
    next_x = (current_x + random_vertex[0]) / 2
    next_y = (current_y + random_vertex[1]) / 2
    next_point = (next_x, next_y)

    # c. Add the new point to the list:
    points.append(next_point)

# 6. Plot the Points:
def plot_sierpinski(points):
    """Plots the Sierpinski triangle points using matplotlib."""
    x_coords = [p[0] for p in points]  # Extract x coordinates
    y_coords = [p[1] for p in points]  # Extract y coordinates

    plt.figure(figsize=(8, 8)) # Make it square, and a bit bigger
    plt.scatter(x_coords[50:], y_coords[50:], s=0.1, color='blue')  # Adjust the points to ignore

    plt.title("Sierpinski Triangle")
    plt.xlabel("X Coordinate")
    plt.ylabel("Y Coordinate")
    plt.axis('equal') # Keep aspect ratio so the triangle is not distorted
    plt.show()

plot_sierpinski(points)
```

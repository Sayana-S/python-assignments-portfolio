```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def normal_density(mean, variance, x):
    sigma = np.sqrt(variance)  # standard deviation
    coefficient = 1 / (np.sqrt(2 * np.pi * variance))
    exponent = np.exp(-((x - mean) ** 2) / (2 * variance))
    return coefficient * exponent

def integration(mean, variance, a, b):
    result, _ = quad(lambda x: normal_density(mean, variance, x), a, b)
    return result

probability = integration(171, 7.1, 162, 190)
print(f"Probability of height between 162 cm and 190 cm: {probability:.4f}")

means = [0, 0, 0]
variances = [0.5, 1, 2]
x_values = np.linspace(-10, 10, 150)

plt.figure(figsize=(5, 5))
for mean, variance in zip(means, variances):
    y_values = [normal_density(mean, variance, x) for x in x_values]
    plt.plot(x_values, y_values, label=f"μ={mean}, σ²={variance}")

plt.title("Normal Density Function")
plt.xlabel("x")
plt.ylabel("fX(x)")
plt.legend()
plt.show()
```
